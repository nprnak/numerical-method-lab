## 1. Linear Regression

### Theory

Linear regression finds the best-fitting straight line through data points by minimizing the sum of squared errors.

**Model Equation:**

$$y = a + bx$$

Where:
- $a$ = y-intercept (value of y when x = 0)
- $b$ = slope (rate of change)

**Objective Function:**

Minimize the sum of squared residuals:

$$S = \sum_{i=1}^{n} (y_i - (a + bx_i))^2$$

**Normal Equations:**

Taking partial derivatives and setting to zero:

$$\frac{\partial S}{\partial a} = -2\sum_{i=1}^{n} (y_i - a - bx_i) = 0$$

$$\frac{\partial S}{\partial b} = -2\sum_{i=1}^{n} x_i(y_i - a - bx_i) = 0$$

This gives:

$$b = \frac{n\sum xy - \sum x \sum y}{n\sum x^2 - (\sum x)^2}$$

$$a = \frac{\sum y - b\sum x}{n}$$

**Algorithm:**

1. Initialize: $sumx = 0$, $sumy = 0$, $sumx2 = 0$, $sumxy = 0$
2. For $i = 1$ to $n$:
   - $sumx = sumx + x_i$
   - $sumy = sumy + y_i$
   - $sumx2 = sumx2 + x_i^2$
   - $sumxy = sumxy + x_i y_i$
3. $denominator = n \cdot sumx2 - (sumx)^2$
4. $b = \frac{n \cdot sumxy - sumx \cdot sumy}{denominator}$
5. $a = \frac{sumy - b \cdot sumx}{n}$
6. Return $(a, b)$

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.array([1,2,3,4,5],dtype=np.float32)
y=np.array([10,8,6,5,4],dtype=np.float32)
n=len(x)
xlen=len(y)
ylen=len(y)
sumx=0
sumx2=0
sumy=0
sumxy=0

if (xlen!=ylen):
  print("Error")
else:
  for i in range(n):
    sumx +=x[i]
    sumy +=y[i]
    sumx2 +=x[i]**2
    sumxy +=x[i]*y[i]

  denominator=n * sumx2 - sumx * sumx

  b = (n * sumxy - sumx * sumy) / denominator
  a = (sumy - b * sumx) / n

  if b<0:
    sign=""
  else:
    sign="+"

  print(f"\nFitted line: y = {a:.2f} {sign} {b:.2f} x")
   # Plot the data points and regression line
  plt.figure(figsize=(8, 6))
  plt.scatter(x, y, color='blue', label='Data points')

  x_line = np.linspace(min(x), max(x), 100)
  y_line = a + b * x_line
  plt.plot(x_line, y_line, color='red', label=f'Regression line y = {a:.2f}{sign}{b:.2f}x')
  plt.legend()
  plt.xlabel('X')
  plt.ylabel('Y')
  plt.title('Linear Regression')
  plt.grid(True)

  plt.show()
```
